## Normalizing Genres

### Getting Unique List of Genres

```python
basics = pd.read_csv("Data/title_basics.csv.gz")
basics.head()
```

```python
## create a col with a list of genres
basics['genres_split'] = basics['genres'].str.split(',')
basics
```

```python
exploded_genres = basics.explode('genres_split')
exploded_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

### Creating Title Genres table

```python
title_genres = exploded_genres[["tconst", "genres_split"]].copy()
title_genres.head()
```

### Creating Genre Mapper Dictionary

To replace string genres with integers

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

### Replacing string genres in title_genres with new integer ids

```python
## make new integer genre_id and drop string genres
#basics['genre_id'] = exploded_genres["genres_split"].map(genre_map)
#basics["genre_id"].replace(genre_map)
#basics = basics.drop(columns='genres_split')
```

```python
basics = exploded_genres

basics["genre_id"] = basics["genres_split"].map(genre_map)
basics = basics.drop(columns = ["genres"])
```

```python
basics.head()
```

```python
#making genres dataframe
genres = pd.DataFrame({"genre_name":genre_map.keys(),
                      "genre_id":genre_map.values()})
genres.head()
```

```python
#making title_genres dataframe
title_genres = pd.DataFrame({"tconst": basics["tconst"],
                            "genre_id": basics["genre_id"]})
title_genres.head()
```

```python
#cleaning basics dataframe
basics.head()
basics = basics.drop(columns = ["originalTitle", "isAdult", "titleType", "genres_split", "genre_id"])
```

## Converting Tables to SQL

Make sure to set a Primary Key for each table that isn't a joiner table (e.g. title_genres is a joiner table).

After creating each table, show the first 5 rows of that table using a SQL query.

Make sure to run the "SHOW TABLES" SQL query at the end of your notebook to show that all required tables have been created.

### Converting title_basics into SQL

Primary Key: tconst (string)

```python
basics.head()
```

```python
basics.duplicated().sum()
```

```python
basics = basics.drop_duplicates()

basics.duplicated().sum()
```

```python
basics.dtypes
```

```python
## Calculate max string lengths for object columns
key_len = basics['tconst'].fillna('').map(len).max()
title_len = basics['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear': Float(),
    'endYear': Float(),
    'runtimeMinutes': Integer()}
```

```python
# Save to sql with dtype and index=False
basics.to_sql('title_basics',engine,dtype = df_schema,if_exists='replace',index=False)
```

```python
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
```

### Converting title_ratings into SQL

Primary Key: titleId (string)

```python
ratings = pd.read_csv("Data/title_akas.csv.gz")
ratings.head()
```

```python
ratings.duplicated().sum()
```

```python
ratings.dtypes
```

```python
## Calculate max string lengths for object columns
key_len = ratings['titleId'].fillna('').map(len).max()
title_len = ratings["title"].fillna("").map(len).max()
region_len = ratings["region"].fillna("").map(len).max()
language_len = ratings["language"].fillna("").map(len).max()
types_len = ratings["types"].fillna("").map(len).max()
attributes_len = ratings["attributes"].fillna("").map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "titleId": Text(key_len+1), 
    "ordering": Integer(),
    "title": String(title_len+1),
    "region": Text(region_len + 1),
    "language": Text(language_len + 1),
    "types": Text(types_len + 1),
    "attributes": Text(attributes_len + 1),
    "isOriginalTitle": Float()
    }
```

```python
#Testing to make sure sql loading happened
q = """
SHOW TABLES;
"""
pd.read_sql(q,engine)
```

```python
# Save to sql with dtype and index=False
ratings.to_sql('title_ratings',engine,dtype = df_schema, if_exists='replace',index=False)
```

```python
q = """
SELECT titleId
FROM title_ratings
LIMIT 5
;
"""
pd.read_sql(q,engine)
```

```python
engine.execute('ALTER TABLE title_ratings ADD PRIMARY KEY (`title`);')
```
